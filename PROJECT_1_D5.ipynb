LINEAR REGRESSION

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
import numpy as np
import pandas as pd
```

```python
apple=pd.read_csv("/content/apple_quality.csv")
apple
```

```python
apple.dropna(inplace=True)
```

```python
r=np.array(apple["Size"]).reshape(-1,1)
u=apple["Sweetness"]
lr.fit(r,u)
```

```python
lr.predict([[2.19]])
```

DECISION TREE

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
V_col=["Acidity","Crunchiness","Juiciness"]
V=apple[V_col]
Target=apple.Quality
```

```python
model=DecisionTreeClassifier()
model.fit(V,Target)
```

```python
a=float(input("Enter Acidity  of the Apple"))
c=float(input("Enter Crunchiness of the Apple"))
d=float(input("Enter Juiciness of the Apple"))
p=model.predict([[a,c,d]])
if p=='good':
  print("Good Quality")
else:
  print("Bad Quality")
```

```python
from sklearn.tree import export_graphviz,plot_tree
```

```python
plot_tree(model)
```

RANDOM FOREST

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
V_cols=['Weight','Crunchiness','Ripeness']
V=apple[V_cols]
Target=apple.Quality
```

```python
rt=RandomForestClassifier(n_estimators=15)
rt.fit(V,Target)
```

```python
a=float(input("Enter Acidity  of the Apple"))
c=float(input("Enter Crunchiness of the Apple"))
d=float(input("Enter Juiciness of the Apple"))
p=model.predict([[a,c,d]])
if p=='good':
  print("Good Quality")
else:
  print("Bad Quality")
```

```python
import os
output_dir ="tree_visualizations"
os.makedirs(output_dir,exist_ok=True)
```

```python
rt.estimators_
```

```python
for i,tree in enumerate(rt.estimators_):
  tree_dot_file=os.path.join(output_dir,f"tree_{i}.dot")
  tree_png_file=os.path.join(output_dir,f"tree_{i}.png")
```

```python
from sklearn.tree import export_graphviz
export_graphviz(tree, out_file=tree_dot_file, feature_names=['Weight','Crunchiness','Ripeness'],
                    class_names=[str(cls) for cls in rt.classes_], filled=True, rounded=True)
```

```python
command=f"dot -Tpng {tree_dot_file} -o {tree_png_file}"
os.system(command)
print(f"Tree{i} visualization saved to {tree_png_file}")
```
